- > What is the maximum duration of a Movie/Show on Netflix ? 

```python
data['Mints']=data['Duration'].str.split(' ').str.get(0)
```

```python
data['Cat_Duration']=data['Duration'].str.split(' ').str.get(1)
```

```python
data.drop(columns='Duration',axis=1,inplace=True)
```

```python
data['Mints']=data['Mints'].astype(int)
```

```python
Top_Lengthy_cat = data.groupby('Category')[['Mints','Cat_Duration']].max().reset_index()
```

```python
plt.figure(figsize=(5,2))
Top_Lengthy_cat.plot(kind='barh')
plt.title('Top_Lengthy_TVShows & Seasons')
plt.xlabel('Mints')
plt.ylabel('Category')
plt.legend()
plt.grid()
plt.show()
```
